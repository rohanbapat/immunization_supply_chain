```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from matplotlib import lines
plt.style.use('grayscale')
```

```python
t = 100
levels = 60
lambdas = [10,20,30,40]
```

```python
I = np.zeros([len(lambdas),levels])
B = np.zeros([len(lambdas),levels])
```

```python
s_levels = [i for i in range(levels)]
```

```python
for idx,l in enumerate(lambdas):
    D = np.random.poisson(lam=l, size=t)
    for s in s_levels:
        I[idx,s] = np.mean(np.maximum([s]*t-D,0))
        B[idx,s] = np.mean(np.maximum(D-[s]*t,0))
```

```python
linestyles = ['-', '--', '-.',':']
```

```python
x = np.arange(levels)
plt.ylim([0, 20])
for idx,_ in enumerate(lambdas):
    plt.plot(x,B[idx,:], linestyles[idx], color = 'black')
    plt.plot(x,I[idx,:], linestyles[idx], color = 'black')
plt.show()
```

```python
lam = 30
```

```python
lamT = 20
A_bar = []
B_bar = []
p_bar = []

for s in range(40):
    A_bar.append(1-stats.poisson.cdf(k=s-1, mu=lamT))
    b_mean = lamT - sum([1-stats.poisson.cdf(k=j, mu = lamT) for j in range(s)])
    B_bar.append(b_mean)
    p_bar.append(min((s+b_mean)/(lamT+b_mean),1))
```

```python
plt.plot(np.arange(len(A_bar)),A_bar,color = 'black')
plt.grid()
plt.show()
```

```python
plt.plot(np.arange(len(B_bar)),B_bar,color = 'black')
plt.grid()
plt.show()
```

```python
plt.plot(np.arange(len(p_bar)),p_bar,color = 'black')
plt.grid()
plt.show()
```
